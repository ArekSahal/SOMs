```python
df.drop([ 'id','key','release_date','mode',"year","name", "explicit"], axis=1, inplace=True
)
df.columns
```

```python
df_artist = pd.read_csv("data_w_genres.csv")
print(df_artist.shape)
df_artist.columns
```

```python
for i in range(500):
    print("index: ", i)
    print(df_artist.artists[i + 5000])
    print(df_artist.genres[i + 5000])
```

```python
genres = ["hip hop", "show tunes", "pop", "classical", "rock", "r&b", "latin"]
artist_by_genre = []
for i in range(len(genres)):
    artist_by_genre.append([])

for i in range(10000,20000):
    df_artist.artists[i]
    for j in range(len(genres)):
        if df_artist.genres[i].find("'"+genres[j]+"'") != -1:
            artist_by_genre[j].append(df_artist.artists[i])

for g in range(len(artist_by_genre)):
    
    print(genres[g], " ",len(artist_by_genre[g]))
            
```

```python
look_through = 5000
"""df = df.sample(n=look_through)
df.shape"""
```

```python
songs_by_genre = []
for i in range(len(genres)):
    songs_by_genre.append([])
    
for i in range(look_through):
    if i % (look_through*0.1)==0:
        print(i)
    for j in range(len(genres)):
        for artist in artist_by_genre[j]:
            if df.artists[i].find(artist) != -1:
                songs_by_genre[j].append(i)

for g in range(len(artist_by_genre)):
    
    print(genres[g], " ",len(songs_by_genre[g]))
    

```

```python
df_clean = df.drop([ 'artists'], axis=1)
df_clean
```

```python
dirty_array = df_clean.to_numpy()
```

```python
dirty_array[0:5, :]
```

```python
labels = []
rows = []
for i in range(len(genres)):
    labels = labels + [i]*len(songs_by_genre[i])
    rows = rows + songs_by_genre[i]
```

```python
data = dirty_array[rows,:]
print(data.shape)
print(len(labels))
```

```python
# Normalize data
data = data - data.mean(axis=0)
data = data / np.abs(data).max(axis=0)
data

```

```python
with open('clean_data.npy', 'wb') as f:
    np.save(f, data)
with open('data_labels.npy', 'wb') as f:
    np.save(f, labels)
with open("data_genres.npy", "wb") as f:
    np.save(f, genres)
```
